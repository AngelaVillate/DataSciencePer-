```python
import re
```

```python
with open("Noticia.txt", encoding='utf-8') as archivo: 
    data = archivo.read()
```

```python
print (data)
```

#### Conversión a minúsculas

```python
data1 = data.lower()
print(data[:500])
```

### preprocesamiento de texto

```python
data2 = re.sub(r"[\W\d_]+", " ", data1)
print(data2[:100])
```

### tokenización

```python
data3 = data2.split()
data3
```

### quitar palabras vacías

```python
from nltk.corpus import stopwords
stopwords_sp = stopwords.words('spanish')
```

```python
data4 = [palabra for palabra in data3 if palabra not in stopwords_sp]
data4
```

### Palabras más usadas

```python
from collections import Counter
Counter(data4).most_common(10)
```

### Steming del documento

```python
from nltk.stem.snowball import SnowballStemmer
spanishStemmer=SnowballStemmer("spanish")
```

```python
data_stem = [spanishStemmer.stem(palabra) for palabra in data4]
data_stem
```

### raíces más usadas

```python
from collections import Counter
Counter(data_stem).most_common(10)
```
